```python
#wap to print left handed triangle.
def left_tri(n):
    for i in range(0,n+1):
        for j in range(i+1):
            print("* ",end='')
        print()

left_tri(5)
```

```python
def right_tri(n):
    for i in range(1,n+1):
        for j in range(n-i):
            print(" ",end="")
        for k in range(i):
            print("*",end="")
        print()

right_tri(5)
```

```python
def centre_tri(n):
    for i in range(1,n+1):
        for j in range(n-i):
            print(" ",end = '')
        for k in range(1,i*2):
            print("*",end = '')
        print()

centre_tri(5)
```

```python
def rev_left(n):
    for i in range(n,0,-1):
        for j in range(i):
            print("*",end = '')
        print()

rev_left(5)
```

```python
def rectangle(n):
    for i in range(1,n+1):
        for j in range(n):
            print("*",end="")
        print()

rectangle(5)
```
